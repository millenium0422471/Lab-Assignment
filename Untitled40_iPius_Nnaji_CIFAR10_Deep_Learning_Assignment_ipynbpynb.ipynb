```python
from tensorflow import keras

(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
import matplotlib.pyplot as plt

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 6))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i])
    plt.title(class_names[int(y_train[i])])
    plt.axis("off")

plt.suptitle("Sample CIFAR-10 Images")
plt.show()
```

```python
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

print("Minimum pixel value:", X_train.min())
print("Maximum pixel value:", X_train.max())
```

```python
y_train = y_train.flatten()
y_test = y_test.flatten()

print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),
    layers.Dense(512, activation="relu"),
    layers.Dense(256, activation="relu"),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])

model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

print("Model compiled successfully.")
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=15,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

print("Test Loss:", round(test_loss, 4))
print("Test Accuracy:", round(test_accuracy, 4))
```

```python
import numpy as np

y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

print("Predictions completed.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="weighted", zero_division=0)
recall = recall_score(y_test, y_pred, average="weighted", zero_division=0)
f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)

print("Model Performance Results")
print("Accuracy:", round(accuracy, 4))
print("Precision:", round(precision, 4))
print("Recall:", round(recall, 4))
print("F1 Score:", round(f1, 4))
```

```python
from sklearn.metrics import classification_report

print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names, zero_division=0))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
from tensorflow import keras
from tensorflow.keras import layers

improved_model = keras.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),
    layers.Dense(1024, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(512, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(256, activation="relu"),
    layers.Dense(10, activation="softmax")
])

improved_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

improved_model.summary()
```

```python
improved_history = improved_model.fit(
    X_train,
    y_train,
    epochs=15,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)
```

```python
=improved_loss, improved_accuracy = improved_model.evaluate(X_test, y_test, verbose=0)

print("Original Model Accuracy:", round(test_accuracy, 4))
print("Improved Model Accuracy:", round(improved_accuracy, 4))
```

```python
=import pandas as pd
import matplotlib.pyplot as plt

comparison_df = pd.DataFrame({
    "Model": ["Original Feedforward NN", "Improved NN with Dropout"],
    "Test Accuracy": [test_accuracy, improved_accuracy]
})

print(comparison_df)

comparison_df.set_index("Model").plot(kind="bar", figsize=(8, 5))

plt.title("Original Model vs Improved Model")
plt.ylabel("Test Accuracy")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.show()
```

## Introduction
This project uses the CIFAR-10 dataset to build a deep learning image classification model. CIFAR-10 contains 60,000 color images divided into 10 categories: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, and truck. The purpose of this project is to show how a neural network can classify everyday objects using image data.

## Dataset Description
The CIFAR-10 dataset contains 50,000 training images and 10,000 testing images. Each image is a 32 by 32 color image with three color channels: red, green, and blue. The dataset was loaded using TensorFlow/Keras. Sample images were visualized to understand the different object categories.

## Methodology
The image pixel values were normalized from the original 0 to 255 range into a 0 to 1 range. This helps the neural network train more effectively. A feedforward neural network was created using a Flatten layer, Dense hidden layers with ReLU activation, and a Softmax output layer for multi-class classification.

The model was compiled using the Adam optimizer, sparse categorical crossentropy loss, and accuracy as the metric. It was trained for 15 epochs using a batch size of 64 and a validation split of 20%.

## Results and Evaluation
The model was evaluated using accuracy, precision, recall, and F1-score. A classification report was generated to show performance across all 10 categories. A confusion matrix was also created to show where the model made correct and incorrect predictions.

Training and validation accuracy and loss graphs were used to monitor the training process. These graphs helped show whether the model improved over time and whether overfitting occurred.

## Model Improvement
An improved neural network was created by increasing the number of neurons and adding Dropout layers. Dropout helps reduce overfitting by randomly ignoring some neurons during training. The original and improved models were compared using a bar chart based on test accuracy.
