```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from fairlearn.datasets import fetch_adult
from fairlearn.metrics import MetricFrame, selection_rate, false_positive_rate, true_positive_rate

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

import shap
from lime.lime_tabular import LimeTabularExplainer
```

```python
# Load Adult Income dataset
adult = fetch_adult(as_frame=True)

X = adult.data
y = adult.target

# Convert target to binary
y = y.astype(str).str.strip()
y = (y == ">50K").astype(int)

print("Dataset shape:", X.shape)
print("Target distribution:")
print(y.value_counts())
X.head()
```

```python
# Sensitive attribute for fairness analysis
sensitive_feature = X["sex"]

print("Sensitive attribute groups:")
print(sensitive_feature.value_counts())
```

```python
# Drop sensitive feature from model inputs
X_model = X.drop(columns=["sex"])

# Identify categorical and numerical columns
categorical_cols = X_model.select_dtypes(include=["object", "category"]).columns.tolist()
numerical_cols = X_model.select_dtypes(include=["int64", "float64"]).columns.tolist()

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
X_train, X_test, y_train, y_test, A_train, A_test = train_test_split(
    X_model,
    y,
    sensitive_feature,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Training size:", X_train.shape)
print("Testing size:", X_test.shape)
```

```python
# Preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
    ]
)
```

```python
# Build model pipeline
model = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("classifier", LogisticRegression(max_iter=1000, class_weight="balanced"))
    ]
)

# Train model
model.fit(X_train, y_train)

print("Model training complete.")
```

```python
# Predictions
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy, 4))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - Logistic Regression")
plt.show()

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Define fairness and performance metrics
metrics = {
    "accuracy": accuracy_score,
    "selection_rate": selection_rate,
    "false_positive_rate": false_positive_rate,
    "true_positive_rate": true_positive_rate
}

# Create MetricFrame
metric_frame = MetricFrame(
    metrics=metrics,
    y_true=y_test,
    y_pred=y_pred,
    sensitive_features=A_test
)

print("Overall Metrics:")
print(metric_frame.overall)

print("\nMetrics by Sensitive Group:")
print(metric_frame.by_group)
```

```python
# Bar plot of fairness metrics by group
metric_frame.by_group.plot(kind="bar", figsize=(10, 6))

plt.title("Fairness Metrics by Sex")
plt.ylabel("Metric Value")
plt.xticks(rotation=0)
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
# Transform training and test data
X_train_transformed = model.named_steps["preprocess"].transform(X_train)
X_test_transformed = model.named_steps["preprocess"].transform(X_test)

# Get feature names
feature_names = model.named_steps["preprocess"].get_feature_names_out()

# Convert to dense if needed
if hasattr(X_train_transformed, "toarray"):
    X_train_transformed = X_train_transformed.toarray()

if hasattr(X_test_transformed, "toarray"):
    X_test_transformed = X_test_transformed.toarray()

# SHAP explainer for logistic regression
classifier = model.named_steps["classifier"]

explainer = shap.LinearExplainer(classifier, X_train_transformed)
shap_values = explainer(X_test_transformed)

# SHAP summary plot
shap.summary_plot(
    shap_values,
    X_test_transformed,
    feature_names=feature_names
)
```

```python
# Explain one individual prediction
sample_index = 0

shap.plots.waterfall(shap_values[sample_index])
```

```python
# Create LIME explainer
lime_explainer = LimeTabularExplainer(
    training_data=np.array(X_train_transformed),
    feature_names=feature_names,
    class_names=["Income <=50K", "Income >50K"],
    mode="classification"
)

# Function for LIME prediction
def lime_predict(data):
    return classifier.predict_proba(data)

# Explain one prediction
lime_exp = lime_explainer.explain_instance(
    data_row=X_test_transformed[0],
    predict_fn=lime_predict,
    num_features=10
)

lime_exp.show_in_notebook(show_table=True)
```

# Ethical AI, Fairness, and Explainability Analysis

## 1. Introduction to Ethical AI
Ethical AI focuses on building machine learning systems that are fair, transparent, accountable, and safe. It is important because machine learning models can unintentionally learn bias from historical data and produce unfair outcomes for certain groups.

## 2. Dataset Description and Preprocessing
This project uses the Adult Income dataset. The goal is to predict whether a person earns more than $50K per year. The sensitive attribute used for fairness analysis is sex. The target variable was converted into binary form, where 1 means income greater than $50K and 0 means income less than or equal to $50K.

Preprocessing included removing the sensitive feature from model training, scaling numerical variables, and one-hot encoding categorical variables.

## 3. Model Architecture and Evaluation
A Logistic Regression model was trained using scikit-learn. The dataset was split into training and testing sets. Model performance was evaluated using accuracy, confusion matrix, precision, recall, F1-score, and classification report.

## 4. Fairness Analysis
Fairness was evaluated using Fairlearn. The metrics included selection rate, false positive rate, and true positive rate across sex groups. These metrics helped identify whether the model performed differently for different groups.

## 5. Explainability Analysis
SHAP was used for global and local explainability. The SHAP summary plot showed which features had the strongest influence on model predictions. A SHAP waterfall plot explained one individual prediction. LIME was also used to explain a single prediction by showing the most important features affecting the result.

## 6. Ethical Considerations and Recommendations
The model may still reflect bias from historical income patterns, even after removing the sensitive attribute from training. Other features may act as proxies for sex or social background. To improve fairness, the organization should monitor fairness metrics regularly, test different preprocessing strategies, consider fairness-aware algorithms, and review model decisions with human oversight.

## 7. Conclusion
The project demonstrated how machine learning can be evaluated not only for accuracy but also for fairness and explainability. Logistic Regression provided a transparent baseline model, while Fairlearn, SHAP, and LIME helped identify group-level fairness issues and explain individual predictions.
